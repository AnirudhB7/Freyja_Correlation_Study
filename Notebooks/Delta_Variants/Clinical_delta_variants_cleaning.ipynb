```python
# Create a date range from 07th Feb 2021 to 30th April 2023
dates = pd.date_range(start='2021-02-06', end='2022-02-26', freq='W-SUN')

# Create a list of MMR values for each date
#mmr = np.random.randint(1000, 8000, size=len(dates))

# Create a dataframe with two columns: date and MMR
dates_df = pd.DataFrame({'date': dates})# 'MMR': mmr})

# Print the dataframe
print(dates_df)
```

```python
path = '/content/OSUCorrelation (1).csv'
```

```python
df = pd.read_csv(path)
df.head()
```

```python
df["Collection date"] = pd.to_datetime(df['Collection date'])
df["Collection date"] = df['Collection date'].dt.strftime('%Y-%m-%d')
df
```

```python
columns_to_drop = [
    "Virus name", "Accession ID", "Location", "Country", "State", "County", "Host",
    "Additional location information", "Sampling strategy", "Gender", "Patient age",
    "Patient status", "Last vaccinated", "Passage", "Specimen", "Additional host information", "Clade",
    "AA Substitutions", "Region", "Counties"
]
df_gisaid = df.drop(columns=columns_to_drop)
df_gisaid
```

```python
df_gisaid = df_gisaid.dropna()
df_gisaid
```

```python
start_date = '2021-02-07'
end_date = '2022-02-26'
```

```python
df_gisaid_filetered = df_gisaid[(df_gisaid['Collection date'] >= start_date) & (df_gisaid['Collection date'] <= end_date)]

df_gisaid_filetered
```

```python
df_gisaid_filetered = df_gisaid_filetered.reset_index(drop=True)
df_gisaid_filetered
```

```python
df_gisaid_filetered['Lineage_num'] = 1
df_gisaid_filetered
```

```python
df_gisaid_filetered_comb = df_gisaid_filetered.join(dates_df['date'])
df_gisaid_filetered_comb
```

```python
import pandas as pd
import numpy as np

# Iterate through the rows of the dataframe
for i in range(len(df_gisaid_filetered_comb)):
    # Check if the current row has a valid date and if it's not the last row
    if not pd.isnull(df_gisaid_filetered_comb['date'].iloc[i]) and i < len(df_gisaid_filetered_comb)-1:
        # Get the current and next dates
        curr_date = df_gisaid_filetered_comb['date'].iloc[i]
        next_date = df_gisaid_filetered_comb['date'].iloc[i+1]

        # Get the sum of frequencies for sample dates that lie between the current and next dates
        freq_sum = 0
        for j in range(len(df_gisaid_filetered_comb)):
            if not pd.isnull(df_gisaid_filetered_comb['Collection date'].iloc[j]):
                sample_date = pd.Timestamp(df_gisaid_filetered_comb['Collection date'].iloc[j])
                if curr_date <= sample_date < next_date:
                    freq_sum += df_gisaid_filetered_comb['Lineage_num'].iloc[j]  # Use 'Lineage_num' instead of 'freq'

        # Store the sum of frequencies in the 'freq' column for the current row
        df_gisaid_filetered_comb.at[i, 'freq'] = freq_sum

print(df_gisaid_filetered_comb)
```

```python
df_gisaid_filetered_comb.to_csv("freyjastudy_gisaid.csv", index = False)
```
