#Script for post run modifications of sample names and stuff

```python
# Standard Imports
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
path = '/content/high_def_run003.xlsx'
```

```python
rundf = pd.read_excel(path)
```

```python
rundf
```

```python


# Function to correct the format
def correct_format(entry):
    parts = entry.split('-')
    if len(parts) == 6:
        year = parts[4]
        if len(year) == 4:
            year = year[-2:]
            parts[4] = year
            return '-'.join(parts)
    return entry

# Remove sample names without "Inf"
rundf = rundf[rundf['sample_name'].str.contains('Inf')]

# Apply the function to the 'sample_name' column to correct the format
rundf['sample_name'] = rundf['sample_name'].apply(correct_format)

# Display the corrected dataframe
rundf
```

```python
path2 = '/content/Combined_aggregated_OHA_normalized_2023-11-08.xlsx'
```

```python
masterdf = pd.read_excel(path2)
masterdf
```

```python
for _, master_row in masterdf.iterrows():
    master_id = master_row['MasterID']
    match = rundf['sample_name'] == master_id
    if match.any():
        rundf.loc[match, 'Location'] = master_row['Location']
        rundf.loc[match, 'date'] = master_row['Sample_Date']
        rundf.loc[match, 'SampleType'] = master_row['SampleType']
        rundf.loc[match, 'Log GC/L'] = master_row['LogCopiesPerL']
rundf
```

```python
rundf['date'] = pd.to_datetime(rundf['date']).dt.strftime('%Y-%m-%d')
rundf
```

```python
rundf.to_excel('high_def_run003_new.xlsx', index = False)
```
