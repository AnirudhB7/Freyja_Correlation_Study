```python
import pandas as pd
import numpy as np

# Iterate through the rows of the dataframe
for i in range(len(df_gisaid_filetered_comb)):
    # Check if the current row has a valid date and if it's not the last row
    if not pd.isnull(df_gisaid_filetered_comb['date'].iloc[i]) and i < len(df_gisaid_filetered_comb)-1:
        # Get the current and next dates
        curr_date = df_gisaid_filetered_comb['date'].iloc[i]
        next_date = df_gisaid_filetered_comb['date'].iloc[i+1]

        # Get the sum of frequencies for sample dates that lie between the current and next dates
        freq_sum = 0
        for j in range(len(df_gisaid_filetered_comb)):
            if not pd.isnull(df_gisaid_filetered_comb['Collection date'].iloc[j]):
                sample_date = pd.Timestamp(df_gisaid_filetered_comb['Collection date'].iloc[j])
                if curr_date <= sample_date < next_date:
                    freq_sum += df_gisaid_filetered_comb['Lineage_num'].iloc[j]  # Use 'Lineage_num' instead of 'freq'

        # Store the sum of frequencies in the 'freq' column for the current row
        df_gisaid_filetered_comb.at[i, 'freq'] = freq_sum

print(df_gisaid_filetered_comb)
```

```python
gisaid_summary['alpha_percentage'] = (gisaid_summary['freq_alpha'] / gisaid_summary['freq'])
gisaid_summary['beta_percentage'] = (gisaid_summary['freq_beta'] / gisaid_summary['freq'])
gisaid_summary['delta_percentage'] = (gisaid_summary['freq_delta'] / gisaid_summary['freq'])
gisaid_summary['epsilon_percentage'] = (gisaid_summary['freq_epsilon'] / gisaid_summary['freq'])
gisaid_summary['gamma_percentage'] = (gisaid_summary['freq_gamma'] / gisaid_summary['freq'])
gisaid_summary['omicron_percentage'] = (gisaid_summary['freq_omicron'] / gisaid_summary['freq'])
```

```python
gisaid_summary
```

```python
gisaid_summary['total_6_variants'] = gisaid_summary['freq_alpha'] + gisaid_summary['freq_beta'] + gisaid_summary['freq_delta'] + gisaid_summary['freq_epsilon'] + gisaid_summary['freq_gamma'] + gisaid_summary['freq_omicron']
gisaid_summary['percentage_of_6_variants'] = (gisaid_summary['total_6_variants'] / gisaid_summary['freq']) * 100
gisaid_summary['other_variants'] = (gisaid_summary['freq'] - gisaid_summary['total_6_variants'])
gisaid_summary

```
