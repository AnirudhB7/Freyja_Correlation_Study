```python
import pandas as pd


siteid_flowdf.rename(columns={
    'Sample_Date': 'date',
    'Log10_GCPerL': 'Log GC/L',
    'Flow_LPerDay': 'Flow_Rate'
}, inplace=True)

# Generalized function to populate missing values in rundf from sitedf
def populate_missing_values(run_df, siteid_flowdf, columns_to_update):
    for index, row in run_df.iterrows():
        # Checking if any specified column has NaN value
        if any(pd.isna(row[col]) for col in columns_to_update):
            master_id = row['sample_name']
            # Search for an exact match in MasterID from site_df
            matched_row = siteid_flowdf[siteid_flowdf['Sample'] == master_id]

            if not matched_row.empty:
                # Update specified columns if available
                for col in columns_to_update:
                    site_col_value = matched_row[col].values[0]
                    if pd.notna(site_col_value):
                        run_df.at[index, col] = site_col_value

# Columns to be updated
columns_to_update = ['Location', 'date', 'SampleType', 'Log GC/L', 'Flow_Rate']

# Populate missing values in rundf using sitedf
populate_missing_values(run_df, siteid_flowdf, columns_to_update)

# Display the updated rundf
run_df
```
